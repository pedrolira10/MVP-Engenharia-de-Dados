# **Catálogo de Dados - Estoque de Chaves Pix por Participante**  
 

---

## **1. Tabela Principal: `fatos_chaves_pix`**  
Contém a quantidade de chaves Pix registradas por instituição, tipo de usuário e tipo de chave.

| Coluna             | Tipo    | Descrição                              | Domínio / Valores Válidos                          |
|--------------------|---------|----------------------------------------|---------------------------------------------------|
| **Data**           | data    | Data de referência do estoque          | Formato: `AAAA-MM-DD`<br>Exemplo: `2023-05-04`    |
| **ISPB**           | texto   | Código ISPB do participante            | 8 dígitos (ex: `00000000`)                        |
| **Nome**           | texto   | Nome da instituição                    | Ex: `Banco do Brasil S.A.`                        |
| **NaturezaUsuario**| texto   | Tipo de usuário                        | `Pessoa Física`<br>`Pessoa Jurídica`              |
| **TipoChave**      | texto   | Tipo de chave registrada               | `CPF`<br>`CNPJ`<br>`Celular`<br>`e-Mail`<br>`Chave Aleatória` |
| **qtdChaves**      | inteiro | Quantidade de chaves ativas            | Mínimo: `0`<br>Máximo: *Variável por instituição* |

---

## **2. Tabelas de Dimensão**  

### **2.1. `dimensao_tempo`**  
Detalhes temporais para análise histórica.

| Coluna | Tipo    | Descrição          | Domínio / Valores Válidos              |
|--------|---------|--------------------|----------------------------------------|
| **Data** | data    | Data completa      | Igual à coluna `Data` em `fatos_chaves_pix` |
| **ano**  | inteiro | Ano (AAAA)         | Ex: `2020`, `2023`                     |
| **mes**  | inteiro | Mês (1-12)         | `1` (Jan) a `12` (Dez)                 |
| **dia**  | inteiro | Dia do mês         | `1` a `31` (conforme mês)              |

---

### **2.2. `dimensao_instituicao`**  
Informações sobre os participantes do Pix.

| Coluna | Tipo   | Descrição               | Domínio / Exemplos                          |
|--------|--------|-------------------------|---------------------------------------------|
| **ISPB** | texto | Código ISPB único       | Ex: `00000000`                              |
| **Nome** | texto | Nome da instituição     | Ex: `Nubank`, `Itaú Unibanco`               |

---

### **2.3. `dimensao_usuario`**  
Classificação dos usuários das chaves.

| Coluna             | Tipo   | Descrição          | Valores Válidos                     |
|--------------------|--------|--------------------|-------------------------------------|
| **NaturezaUsuario**| texto  | Tipo de usuário    | `Pessoa Física`<br>`Pessoa Jurídica` |

---

### **2.4. `dimensao_tipo_chave`**  
Tipos de chaves Pix disponíveis.

| Coluna       | Tipo   | Descrição          | Valores Válidos                                     |
|--------------|--------|--------------------|----------------------------------------------------|
| **TipoChave**| texto  | Categoria da chave | `CPF`<br>`CNPJ`<br>`Celular`<br>`e-Mail`<br>`Chave Aleatória` |

---

## **3. Parâmetros de Consulta**  
Filtros e ordenação para acesso aos dados.

| Parâmetro  | Tipo     | Exemplo de Uso                          | Descrição                             |
|------------|----------|-----------------------------------------|---------------------------------------|
| **$filter**  | texto   | `Data eq '2023-01-01'`                 | Filtra por data específica            |
|            |          | `TipoChave eq 'Celular'`               | Chaves do tipo celular                |
| **$orderby** | texto   | `qtdChaves desc`                       | Ordena por volume de chaves (decrescente) |
| **$top**     | inteiro | `$top=10`                              | Limita a 10 resultados                |
| **$skip**    | inteiro | `$skip=5`                              | Pula os 5 primeiros registros         |
